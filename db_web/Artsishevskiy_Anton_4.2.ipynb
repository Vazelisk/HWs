```python
import requests
from bs4 import BeautifulSoup
import re
from pyquery import PyQuery as pq
from tqdm import tqdm_notebook
import pandas as pd
```

### Задание 1
Определю супчик

```python
stihi_ru = 'https://stihi.ru'
result = requests.get(stihi_ru)
html = result.text
```

Найду строчку кода с ссылкой на произведения

```python
soup = BeautifulSoup(html,'html.parser')
post = soup.find('a', {'class': 'topref'})
match = post.prettify()
print(match)
```

Выделю ссылку и скреплю её

```python
link = re.search('href="(.+?)">', match)
link = link.group(1)
```

```python
link = stihi_ru + link
print(link)
```

### Задание 2
Перезадам супчик

```python
result1 = requests.get(link)
html1 = result1.text
soup = BeautifulSoup(html1,'html.parser')
```

Создам список для названий и заполню его

```python
titles = []
for title in soup.find_all('a', {'class': 'poemlink'}):
    titles.append(title.get_text())
print(titles)
```

Создам список для авторов и заполню его

```python
authors = []
for author in soup.find_all('a', {'class': 'authorlink'}):
    authors.append(author.get_text())
print(authors)
```

Создам список для дат и заполню его

```python
dates = []
for date in soup.find_all('small'):
    date = date.get_text()
    date = re.sub('- ', '', date)
    dates.append(date)
print(sorted(dates))
```

Создам словарь со структурой {'Title': ..все названия.., 'Author': .все авторы..., 'Date': .все даты..}

```python
total = {'Title': [], 'Author': [], 'Date': []}
n = 0
for i in range(len(titles)):
    total['Title'] = total['Title'] + [titles[n]]
    total['Author'] = total['Author'] + [authors[n]]
    total['Date'] = total['Date'] + [dates[n]]
    n +=1
```

Переделаю его в дф

```python
df = pd.DataFrame(total)
df
```

Рассортирую словарь по дате (он сортирует и время так же, но вроде это не противоречит условию)

```python
df.sort_values("Date", axis = 0, ascending = True, 
                 inplace = True, na_position ='first')
df
```
